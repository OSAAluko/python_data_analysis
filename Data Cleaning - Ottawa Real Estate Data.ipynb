```python
# Add the corrected_price column to the dataframe
data_df = pd.concat((data_df, corrected_price), axis = 1)
```

```python
# Renaming the columns and creating a checkpoint...
data_df.columns = ['latitude', 'longitude', 'postalCode', 'walkScore', 'propertyType', 'style', 'yearBuilt', 'no_of_bedrooms', 'no_of_bathrooms', 'parking', 'garage', 'lotDepth', 'lotFrontage', 'price1', 'price2', 'price3', 'corrected_price']
data_df_checkPoint1 = data_df.copy()
data_df
```

```python
# Dropping rows with nans on the corrected_price column
data_df = data_df[data_df["corrected_price"].notna()]
data_df
```

```python
# Checking to confirm that all rows with nans on the corrected_price column are dropped
data_df["corrected_price"].isna().sum()
```

```python
# Dropping columns not needed for the correllation analysis
data_df = data_df.drop(['price1', 'price2', 'price3'], axis = 1)
data_df
```

```python
# Dropping columns not needed for the correllation analysis
data_df = data_df.drop(['latitude', 'longitude', 'postalCode', 'walkScore'], axis = 1)
data_df
```

```python
# Dropping columns not needed for the correllation analysis
data_df = data_df.drop(['parking', 'garage', 'lotDepth', 'lotFrontage'], axis = 1)
data_df
```

```python
#Changing data types for colunms - yearBuilt, no_of_bedrooms, no_of_bathrooms and corrected_price

data_df['yearBuilt'] = pd.to_numeric(data_df['yearBuilt'])

data_df['no_of_bedrooms'] = pd.to_numeric(data_df['no_of_bedrooms'])

data_df['no_of_bathrooms'] = pd.to_numeric(data_df['no_of_bathrooms'])

data_df['corrected_price'] = data_df['corrected_price'].str.replace('"', '')
data_df['corrected_price'] = data_df['corrected_price'].astype(float)
data_df['corrected_price'] = data_df['corrected_price'].astype(int)

data_df
```

```python
# Further categorizing the propertyType column
# nan to 'Unstated'
# 'Vacant Land' as is
# 'Single Family' and 'Multifamily' to 'Residential'
# 'Residential' as is
# 'Commercial' and 'Retail' to 'Business'
# 'Farm' and 'Lots/Acreage' to 'Farm Land'

data_df['propertyType'] = data_df['propertyType'].fillna('Not Stated')

data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Single Family', value = 'Residential')
data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Multifamily', value = 'Residential')
data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Commercial', value = 'Business')
data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Retail', value = 'Business')
data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Farm', value = 'Farm Land')
data_df['propertyType'] = data_df['propertyType'].replace(to_replace = 'Lots/Acreage', value = 'Farm Land')


data_df['propertyType'].unique()

```

```python
# Further categorizing the style column
# nan to 'Unstated'
# 'Bungalow' as is
# '1 1/2 Storey', '2-Storey', 'Loft' and 'Multi-Level' to 'Storey/Loft/Multi-Level'
# 'Detached' and 'Semi Detached' to 'Detached/Semi-Detached'
# 'Apartment', 'Stacked' and 'Row Unit' to 'Apartments'

data_df['style'] = data_df['style'].fillna('Not Stated')

data_df['style'] = data_df['style'].replace(to_replace = '1 1/2 Storey', value = 'Storey/Loft/Multi-Level')
data_df['style'] = data_df['style'].replace(to_replace = '2-Storey', value = 'Storey/Loft/Multi-Level')
data_df['style'] = data_df['style'].replace(to_replace = 'Loft', value = 'Storey/Loft/Multi-Level')
data_df['style'] = data_df['style'].replace(to_replace = 'Multi-Level', value = 'Storey/Loft/Multi-Level')
data_df['style'] = data_df['style'].replace(to_replace = 'Detached', value = 'Detached/Semi-Detached')
data_df['style'] = data_df['style'].replace(to_replace = 'Semi Detached', value = 'Detached/Semi-Detached')
data_df['style'] = data_df['style'].replace(to_replace = 'Apartment', value = 'Apartments')
data_df['style'] = data_df['style'].replace(to_replace = 'Stacked', value = 'Apartments')
data_df['style'] = data_df['style'].replace(to_replace = 'Row Unit', value = 'Apartments')

data_df['style'].unique()
```

```python
# Using Numpy
import numpy as np

#Avoiding scientific printouts
np.set_printoptions(suppress = True, linewidth = 100, precision = 2)

# Function for creating data checkpoints wth Numpy
def createCheckPoint(fileName, cHeader, cData):
   np.savez(fileName, header = cHeader, data = cData) 
   checkPointVar = np.load(fileName + '.npz')
   return checkPointVar

```

```python
# Load the pre-cleaned data with Numpy and create a initial data checkpoint
cleanedData = np.genfromtxt('cleaned_data.csv', delimiter=',', autostrip = True)
createCheckPoint('cleanedDataCheckPoint1', cleanedData[0, :], cleanedData)
```

```python
# Reload data **without header**
preprocessingData = np.genfromtxt('cleaned_data.csv', delimiter=',', autostrip = True, skip_header = 1)
preprocessingData
```

```python
#Get a temporary filler for nans
temporaryFiller = np.nanmax(preprocessingData).round(2) + 1000
temporaryFiller
```

```python
# Saving initial Data Stats - Min Mean & Max - irrespective of the nan values
initialMin = np.nanmin(preprocessingData, axis = 0).round(2)
initialMeans = np.nanmean(preprocessingData, axis = 0)
initialMax = np.nanmax(preprocessingData, axis = 0).round(2)
initial_stats = np.array([initialMin, initialMeans, initialMax], dtype = object)
initial_stats
```

```python
# Get indices of string columns and numeric columns
stringsColumnIndices = np.argwhere(np.isnan(initialMeans)).squeeze()
numericColumnIndices = np.argwhere(np.isnan(initialMeans) == False).squeeze()
```

```python
# Reload string data seperately from numeric data
stringData = np.genfromtxt('cleaned_data.csv', delimiter=',', skip_header = 1, autostrip = True, usecols = stringsColumnIndices, dtype = str)
stringData
```

```python
# Reload numeric data seperately from string data and replacing nans with the temporary iller
numericData = np.genfromtxt('cleaned_data.csv', delimiter=',', skip_header = 1, autostrip = True, usecols = numericColumnIndices)
# np.unique(numericData[:, 1])
# Saving initial Data Stats - Min Mean & Max - irrespective of the nan values
numericMin = np.nanmin(numericData, axis = 0).round(2)
numericMeans = np.nanmean(numericData, axis = 0).round()
numericMax = np.nanmax(numericData, axis = 0).round(2)
numericStats = np.array([numericMin, numericMeans, numericMax], dtype = object)
numericStats
```

```python
# Reload numeric data seperately from string data and replacing nans with the temporary filler
numericData = np.genfromtxt('cleaned_data.csv', filling_values = temporaryFiller, delimiter=',', skip_header = 1, autostrip = True, usecols = numericColumnIndices)
numericData
```
